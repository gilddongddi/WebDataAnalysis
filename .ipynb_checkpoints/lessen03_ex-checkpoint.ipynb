URL을 이용한 데이터 분석

```python
import requests
```

```python
from bs4 import BeautifulSoup
url= "https://home.sjcu.ac.kr/ko/department/software-engineering-major-department.do"
page = requests.get(url)  
html = page.text
html
```

```python
soup = BeautifulSoup(html, "lxml")
```
