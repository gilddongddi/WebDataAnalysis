# 나무위키 데이터 가져오기

```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome('/opt/homebrew/bin/chromedriver')
```

```python
url = 'https://www.namu.wiki/RecentChanges'
```

```python
driver.get(url)
```

```python
from bs4 import BeautifulSoup
req = driver.page_source
```

```python
soup = BeautifulSoup(req, 'lxml')
```

```python
soup
```

```python
contents_table = soup.find('table')
```

```python
contents_table
```

```python
table_body = contents_table.find('tbody')
```

```python
table_body
```

```python
table_rows = table_body.find_all('tr')
```

```python
table_rows
```

```python
page_url_base = 'https://namu.wiki'
page_urls =[]
for i in range(0, len(table_rows)):
    first_td = table_rows[i].find_all('td')[0]
    td_url = first_td.find_all('a')
    if (len(td_url)>0):
        page_url = page_url_base + td_url[0].get('href')
        if "png" not in page_url:
            page_urls.append(page_url)
```

```python
page_urls = list(set(page_urls))
```

```python
for page in page_urls[:3]:
    print(page)
```

```python
driver.get(page_urls[0])
```

```python
req = driver.page_source
```

```python
soup = BeautifulSoup(req, 'lxml')

soup
```

```python
contents_table = soup.find('article')
```

```python
title = contents_table.find_all('h1')[0]
```

```python
category = contents_table.find_all('ul')[0]
```

```python
content_paragraphs = contents_table.find_all('div')

content_corpus_list = []
for paragraphs in content_paragraphs:
    content_corpus_list.append(paragraphs.text)
content_corpus = "".join(content_corpus_list)
```

```python
print(title.text)
print("\n")
print(category.text)
print("\n")
print(content_corpus)
```
