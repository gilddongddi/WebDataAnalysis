# 12개의 파일을 하나의 리스트로 만들기

```python
import glob
```

```python
files = glob.glob('서울특별시 공공자전거 이용정보(월별)/서울특별시 공공자전거 이용정보(월별)_21.*.csv')
```

```python
files
```

```python
file_all = []
for file in files:
    data_file = pd.read_csv(file, encoding='cp949')
    file_all.append(data_file)
```

```python
file_all
```

```python
data_concat = pd.concat(file_all, axis=0, ignore_index=True)
```

```python
data_concat
```

### 이용건수가 가장 많은 대여소명은?

```python
data_concat.sort_values(by='이용건수', ascending=False)
```
