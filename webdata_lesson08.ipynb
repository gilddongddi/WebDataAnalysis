```python
state_data = pd.read_csv('서울교통공사_1_8호선 역사 좌표(위경도) 정보_20211231.csv', encoding='cp949')
```

```python
state_data.head()
```

```python
map_state=folium.Map(location=[37.5535253, 127.0704573], zoom_start=15)
```

```python
for i, state in state_data.iterrows():
    if state['호선']==1:
        folium.Marker(location=[state['위도'], state['경도']], popup=state['역명'], icon=folium.Icon(color='#0F218B', icon='star')).add_to(map_state)
```

```python
for i, state in state_data.iterrows():
    if state['호선']==2:
        folium.Marker(location=[state['위도'], state['경도']], popup=state['역명'], icon=folium.Icon(color='green', icon='star')).add_to(map_state)
```

```python
for i, state in state_data.iterrows():
    if state['호선']==3:
        folium.Marker(location=[state['위도'], state['경도']], popup=state['역명'], icon=folium.Icon(color='blue', icon='star')).add_to(map_state)
```
