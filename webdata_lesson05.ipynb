```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
```

```python
driver = webdriver.Chrome('/opt/homebrew/bin/chromedriver')
```

```python
url='https://home.sjcu.ac.kr/ko/department/department-software-guide.do'
driver.get(url)
```

```python
from bs4 import BeautifulSoup
```

```python
html=driver.page_source
```

```python
print(html)
```

```python
soup=BeautifulSoup(html, 'lxml')
```

```python
soup
```

```python
tb = soup.find('tbody', {'id':'tbody'})
```

```python
lists = tb.find_all('a', {'class':'dept-layer-pop'})
```

```python
lists
```

```python
result = []
for index in range(0, len(lists)):
    result.append(lists[index].string)
```

```python
result
```

```python
e_list =[]
o_list = []
for i in range(len(result)):
    if (i %2==0):
        e_list.append(result[i])
    else:
        o_list.append(result[i])
```

```python
e_list
```

```python
o_list
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(e_list, columns=['1학기'])
```

```python
df1
```

```python
df2 = pd.DataFrame(o_list, columns=['2학기'])
```

```python
df2
```

```python
DataSet = list(zip(e_list, o_list))
```

```python
df = pd.DataFrame(DataSet, columns = ['1학기', '2학기'])
```
